```python
import pandas as pd
from sklearn.preprocessing import OneHotEncoder
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
```

```python
# Read the data from the CSV file
df = pd.read_csv('Unemployment in India.csv')
```

```python
# Drop rows with null values
df = df.dropna()
```

```python
# Encode categorical columns using OneHotEncoder
encoder = OneHotEncoder()
encoded_data1 = encoder.fit_transform(df[['Region']]).toarray()
encoded_data2 = encoder.fit_transform(df[[' Date']]).toarray()
encoded_data3 = encoder.fit_transform(df[[' Frequency']]).toarray()
encoded_data4 = encoder.fit_transform(df[['Area']]).toarray()
```

```python
df_combined = pd.concat([pd.DataFrame(encoded_data1),
                         pd.DataFrame(encoded_data2),
                         pd.DataFrame(encoded_data3),
                         pd.DataFrame(encoded_data4)],
                        axis=1)
```

```python
df = df.drop(['Region',' Date',' Frequency','Area'],axis = 1)
```

```python
df_final = pd.concat([df,df_combined],axis = 1)
```

```python
df_final
```

```python
df_final.isna().sum()
```

```python
df_final = df_final.dropna()
```

```python
df_final.dtypes
```

```python
X = df_final.drop([' Estimated Unemployment Rate (%)'],axis = 1)
y = df_final[' Estimated Unemployment Rate (%)']
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size = 0.09,random_state = 20)

rfr = RandomForestRegressor()
rfr.fit(X_train,y_train)
y_pred = rfr.predict(X_test)
```

```python
from sklearn.metrics import mean_squared_error,mean_absolute_error
mse = mean_squared_error(y_test,y_pred)
mae = mean_absolute_error(y_test,y_pred)
print('mean squared error is : ',mse)
print('mean absolute error is : ',mae)
```
