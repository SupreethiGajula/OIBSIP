```python
import pandas as pd
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
#we will use k nearest neighbors for this classification
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
```

```python
iris = load_iris()
df = pd.DataFrame(iris.data,columns = iris.feature_names)
```

```python
df
```

```python
df['species'] = pd.Categorical.from_codes(iris.target,iris.target_names)
#the mapping between the target and target_names takes place
#targets are integers and basically our y values
#target names are virginica,setosa and versicolor
```

```python
X = df.drop('species',axis = 1)
y = df['species']
```

```python
#we split the data into test and train
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size = 0.2,random_state = 40)
```

```python
knn = KNeighborsClassifier(n_neighbors = 4)
```

```python
knn.fit(X_train,y_train)
```

```python
#we are finding the predictions with the model we made
y_predicted = knn.predict(X_test)
print(y_predicted)
```

```python
#checking for the accuracy
acc = accuracy_score(y_test,y_predicted)
```

```python
print('accuracy score:',acc)
```
